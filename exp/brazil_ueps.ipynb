```python
import pandas as pd
import xarray
import numpy as np
```

```python
ueps = xarray.open_dataset('UEPS_v1.nc')
ueps
```

```python
ueps = ueps.isel(Turbine=2)
ueps
```

```python
ueps = ueps.to_dataframe()
ueps
```

```python
ueps.info()
```

```python
ueps = ueps.reset_index(["Height","Range"])
```

```python
#flatten_multiindex_df(ueps,["Range","Height"],[features_with_range_dim,features_with_height_dim],{"Range":12,"Height":50})

#def flatten_multiindex_df(df,dimensions,list_of_features_of_dims,split_points_of_dims):
#  df_dims = pd.DataFrame()

  # For every dimension
 # for dim,features_with_dim in zip(dimensions,list_of_features_of_dims):
 #   # For every split point
 #   for i,split_point in enumerate(split_points_of_dims[dim]):
 #     if i < len(split_points_of_dims[dim]):
 #       df_dim1= df.where(df[dim] < split_point and ... )[features_with_dim]
 #     else:
 #       df_dim1= df.where(df[dim] >= split_point)[features_with_dim]
 #     df_dim1 = df_dim1.groupby(level=0).mean()
 #     df_dim1.columns = [column_name + f"{i}" for column_name in features_with_dim]
#
 #     df_dims = pd.merge(df_dims, df_dim1,eft_index=True, right_index=True)

# List of features that have multiple dimensions 
# e.g. range and time
features_with_range_dim = ["lidar_wind_speed","lidar_wind_direction","lidar_wind_speed_std","lidar_ws_u",
                           "lidar_ws_v","lidar_ws_w","lidar_availability"]
# height and time
features_with_height_dim = ["wind_speed","wind_direction","wind_speed_std","wind_direction_std","wind_speed_max",
                            "wind_speed_min","wind_speed_cube","air_temperature", "relative_humidity",
                            "UST","UST_flag","HS","HS_flag","TKE","LMO"]


# Two buckets of range measurements
ueps_range1= ueps.where(ueps["Range"] < 12)[features_with_range_dim]
ueps_range2 = ueps.where(ueps["Range"] >= 12)[features_with_range_dim]
ueps_range1 = ueps_range1.groupby(level=0).mean()
ueps_range2 = ueps_range2.groupby(level=0).mean()

# Rename columns
ueps_range1.columns = [column_name + "1" for column_name in features_with_range_dim]
ueps_range2.columns = [column_name + "2" for column_name in features_with_range_dim]

# Two buckets of height measurments
ueps_height1 = ueps.where(ueps["Height"] < 50)[features_with_height_dim]
ueps_height2 = ueps.where(ueps["Height"] >= 50)[features_with_height_dim]
ueps_height1 = ueps_height1.groupby(level=0).mean()
ueps_height2 = ueps_height2.groupby(level=0).mean()

#Rename columns
ueps_height1.columns = [column_name + "1" for column_name in features_with_height_dim]
ueps_height2.columns = [column_name + "2" for column_name in features_with_height_dim]

#
ueps_rest = ueps.groupby(level=0).mean()
```

```python
features_without_height_and_range_dim = list(set(list(ueps.columns)) 
                                         - (set(features_with_height_dim).union(set(features_with_range_dim)))
                                         )
ueps_rest = ueps_rest[features_without_height_and_range_dim]


df_range = pd.merge(ueps_range1,ueps_range2, left_index=True, right_index=True)
df_height = pd.merge(ueps_height1,ueps_height2, left_index=True, right_index=True)
df_combined1 = pd.merge(ueps_rest, df_range, left_index=True, right_index=True)
df_combined = pd.merge(df_combined1, df_height, left_index=True, right_index=True).drop(columns=["Range","range","Height","Turbine"])

# Assert that no columns / rows have been lost
assert(ueps.shape[0] / 6 / 26 == df_combined.shape[0])
assert(ueps.shape[1]- 4 == df_combined.shape[1] - len(features_with_height_dim) - len(features_with_range_dim))

df_combined
```
